### 9.1 공간 주파수의 이해

주파수 : 이벤트가 주기적으로 재발생하는 빈도



```python
# 9.1.1. 주파수 그리기
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0, 1, 0.001)
Hz = [1, 2, 10, 100]
gs = [np.sin(2 * np.pi * t * h) for h in Hz] # Hz에 따른 sin 파형 생성

plt.figure(figsize=(10, 5))
for i, g in enumerate(gs):
    # 그래프 그리기
    plt.subplot(2, 2, i+1) 
    plt.plot(t, g)
    plt.xlim(0, 1) # x축 범위 지정
    plt.ylim(-1, 1)
    plt.title("frequency: %3d Hz" % Hz[i])
plt.tight_layout()
plt.show()
```

저주파 영역 : 화소 밝기가 거의 변화가 없거나 점직적으로 변화 (배경, 객체 내부)

고주파 영역 : 화소 밝기가 급변하게 변화 (객체 모서리)
